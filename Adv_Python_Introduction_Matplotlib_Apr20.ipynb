```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Companys performance at the year ending 19-20')
plt.xlabel('sales')
plt.ylabel('Revenue')
plt.legend('This is comparsion line')
plt.show()
```

```python
z= [10,5,0]
```

```python
plt.plot(x,y)
plt.plot (x,z)
plt.title('Companys performance at the year ending19-20')
plt.xlabel('sales')
plt.ylabel('Revenue')
plt.legend(['This is x and y', 'this is x and z'])
plt.show()
```
